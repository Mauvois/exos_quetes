# Diabetese Detection Models

This [dataset](https://raw.githubusercontent.com/mansont/datasets-tests/main/diabetese.csv) contains patient data and their diabetese condition: "1" they have diabetes, "0" they do not have diabetese.


Build the following models and compare their performance:
* A logistic regression model
* A single-layer perceptron model
* A multilayer perceptron

## 3 models comparison

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron, LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV
import numpy as np
import pandas as pd
import time
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('raw.githubusercontent.com_mansont_datasets-tests_main_diabetese.csv')
```

```python
X = df[['age', 'pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi', 'dpf']]
y = df['diabetes']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
sc = StandardScaler()
sc.fit(X_train)
StandardScaler(copy=True, with_mean=True, with_std=True)
```

```python
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
models = [Perceptron(), LogisticRegression(max_iter = 500), MLPClassifier(max_iter=3000)]
```

```python
for model in models:
    start = time.time()
    model.fit(X_train_std, y_train)
    end = time.time()
    y_pred = model.predict(X_test_std)
    print(type(model).__name__)
    print("")
    print("Time : {:.4f} sec".format(end-start))
    print("Train Score : {:.3f} , Test Score : {:.3f}".format(model.score(X_train_std,y_train), model.score(X_test_std,y_test)))
    print("Accuracy : {:.3f}".format(accuracy_score(y_test, y_pred)))
    print(30*"°")
```

Perceptron and Logistic regression have very similar results. Though, Logistic regression offers better performance while being only little less fast. MLPClassifier on the other hand seems to be overfitting and is clearly slower than the other 2 algorithms.

#### Is there a notable difference in the MLP performance when a ReLU, Sigmoid or SoftMax activation function is used?

```python
activations = ['relu', 'logistic', 'tanh']
```

```python
params_relu = {
    'hidden_layer_sizes': [
        (50,), (75,), (100,), (125,), (150,),  # One-layer architectures
        (200,), (250,),                        # Larger single-layer architectures
        
        (50, 50), (75, 75), (100, 100), (125, 125),  # Two-layer architectures
        (150, 150), (200, 100),                   # Wider two-layer architectures
        
        (50, 25), (75, 50), (100, 75), (125, 100),  # Asymmetric two-layer architectures
        
        (50, 50, 50), (100, 75, 50),             # Three-layer architectures
        (150, 100, 50), (100, 100, 50),          # Deeper three-layer architectures
    ],
    'alpha': [0.00005],           
    'learning_rate_init': [0.0005], 
}

params_logistic = {
    'hidden_layer_sizes': [
        (30,), (50,), (70,),             # One-layer architectures
        (90,),                           # Adding larger single-layer architecture
        
        (30, 30), (50, 50), (70, 70),    # Two-layer architectures
        (90, 45),                        # Larger two-layer architecture
        
        (50, 25), (70, 35),              # Asymmetric two-layer architectures
        
        (30, 30, 30), (50, 25, 25),      # Three-layer architectures
        (70, 35, 20),                    # Larger three-layer architecture
    ],
    'alpha': [0.00005],   
    'learning_rate_init': [0.001],   
}

params_tanh = {
    'hidden_layer_sizes': [
        (30,), (50,), (70,),             # One-layer architectures
        (90,),                           # Adding larger single-layer architecture
        
        (30, 30), (50, 50), (70, 70),    # Two-layer architectures
        (90, 45),                        # Larger two-layer architecture
        
        (50, 25), (70, 35),              # Asymmetric two-layer architectures
        
        (30, 30, 30), (50, 25, 25),      # Three-layer architectures
        (70, 35, 20),                    # Larger three-layer architecture
    ],
    'alpha': [0.00005],   
    'learning_rate_init': [0.0005],  
}
```

```python
param_grids = [params_relu, params_logistic, params_tanh]
```

```python
models = [MLPClassifier(activation=act, max_iter=1500) for act in activations]
```

```python
results_whole = []

for model, param_grid in zip(models, param_grids):
    start = time.time()
    warnings.filterwarnings("ignore")

    # Perform GridSearchCV
    grid_search = GridSearchCV(model, param_grid, cv=3, scoring='accuracy')
    grid_search.fit(X_train_std, y_train)

    end = time.time()

    print(type(model).__name__ + " with " + model.activation + " activation")
    print("Time: {:.4f} sec".format(end - start))
    print('')
    print("Best Parameters:", grid_search.best_params_)
    print("Best Accuracy:", grid_search.best_score_)

    best_model = grid_search.best_estimator_
    print("Train Score: {:.3f}, Test Score: {:.3f}".format(
        best_model.score(X_train_std, y_train), best_model.score(X_test_std, y_test)))

    y_pred = best_model.predict(X_test_std)
    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy: {:.3f}".format(accuracy))

    results_df = pd.DataFrame(grid_search.cv_results_)
    results_df['model'] = type(model).__name__ + " with " + model.activation + " activation"
    results_whole.append(results_df)

    print(40 * "—")

final_results = pd.concat(results_whole, ignore_index=True)
```

```python
final_results['number_of_layers'] = final_results['param_hidden_layer_sizes'].apply(lambda x: len(x))
final_results = final_results[
    ['model', 'number_of_layers', 'param_hidden_layer_sizes', 
    'mean_fit_time', 'mean_score_time', 'mean_test_score', 'rank_test_score' ]
  ]
```

```python
final_results.sort_values(by='rank_test_score', ascending=True)
```

We can observe that the Tanh activation function achieved the highest mean test score, followed closely by the Logistic activation. The ReLU activation had slightly inferior performance in comparison to the other two. However, these differences are pretty close, and depending on the application, they might not be considered "notable".

### Does the network performance change when the density (number of neurons) of the hidden layers change?

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="model", y="mean_test_score", data=final_results)
plt.title('Mean Test Score by Activation Function')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

ReLU: It's not always the case that increasing the number of neurons or layers enhances the performance. The top performer for a single layer had 125 neurons.

Logistic : There's a slight trend of improved performance with added complexity (layers/neurons), with the (70,35,20) configuration outperforming others.

Tanh : Simpler configurations seem to perform better, particularly the single-layered architecture with 30 neurons.

```python
final_results['total_neurons'] = final_results['param_hidden_layer_sizes'].apply(lambda x: sum(x))
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="number_of_layers", y="mean_test_score", data=final_results)
plt.title('Mean Test Score by Number of Layers')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x="number_of_layers", y="mean_test_score", hue="model", data=final_results, palette="Set2")
plt.title('Mean Test Score by Number of Layers and Activation Function')
plt.legend(title="Activation Function")
plt.tight_layout()
plt.show()
```

The benefits of adding more layers seem to be activation function-specific. While Tanh and ReLU seem to favor simpler models, Logistic might benefit from slightly more complex architectures.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_neurons", y="mean_test_score", hue="model", data=final_results)
plt.title('Mean Test Score by Total Neurons in Hidden Layers')
plt.tight_layout()
plt.show()
```

Increasing the number of neurons in the hidden layer doesn't consistently lead to better performance across activation functions.

For ReLU, adding more layers didn't necessarily result in better performance.

For Logistic, more complex configurations (in terms of layers) had better performance.
For Tanh, simpler configurations with fewer layers performed best.
